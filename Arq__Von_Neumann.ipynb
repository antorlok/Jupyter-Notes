# Arquitectura de Von Neumann.

## Historia y conceptualización:

- John Von Neumann fue el encargado de colaborar entre 1944 y 1946 sobre informes de las primeras computadoras electrónicas.👨‍💻
- Se involucro con la computadora ENIAC, pero después decidió definir un nuevo sistema lógico de computación, basado en ideas de Turing.🧠
  
### La propuesta de Von Neuman en cuanto a arquitectura fue la siguiente :

- Una unidad de control.🚦
  
- Una unidad aritemética y lógica (ALU).🧮

  
- Una unidad de memoria.💾

  
- Registros.📑

  
- Entradas y salidad.🔌

Esta arquitectura se basa en el principio de "Programa almacenado", es decir, que tanto los datos de instrucción como los de el programa, se almacenan en una misma memoria!!🤯✨


## Componentes

#### Unidad Central de Procesamiento(CPU):🧠

Es el encargado de ejecutar las instrucciones de un programa, contiene a la ALU y variedad de registros para poder procesar datos.⚙️

**Para entender estos registros, estos se dividen en:**

- El registro de dirección de memoria(RI): Contiene la dirección de memoria de los datos que necesitan ser accedidos.📍
  
- Registro de datos de memoria(MDR): Contienen datos que estan siendo transferidos hacia o desde la memoria.🔄
  
- Contador de programa(PC): Contiene la dirección de la proxima instrucción a ser ejecutada.➡️
  
- Acumulador(AC): Almacena resultados aritméticos y lógicos intermedios.Σ
 
- Registro de instrucción actual(IR): Contiene la instrucción actual durante el procesamiento.📝
  

Dentro de el CPU encontramos la unidad de control(CU), se podría decir que esta es el cerebro, controla en funcionamiento de la ALU, la memoria y la entrada y salida de la computadora.🚦

**La CU itera funciones vitales como:**
- Secuenciador: Asegura que las instrucciones se ejecuten en orden.🔢
- Decodificador: Traduce las instrucciones en señales electricas.⚡
- Reloj: Sincroniza las operaciones.⏱️

### Los Buses:🚌

Son los medios por los cuales se transmiten los datos, conectan los componentes.

**Se calsifican en:**

- Direcciones: Lleva los direcciones de datos pero no los datos en si.🗺️

- Bus de datos: Lleva datos entre el procesador, memoria y dispositivos de entrada y salida.📦

- Bus de control: Lleva señales de control, comandos desde la cpu.📢

### Memoria:💾

La unidad de memoria consta de ram, referida como memoria principal, a diferencia de un disco duro, esta es rápida y accesible desde la CPU.🚀

**Características:**

- La ram cuenta con distintas direcciones de memoria para cada uno de los registro.🏷️
  
- Cada ubicación de memoria tiene una dirección única.🆔
  
- Se organizan las direcciones en una tabla de direcciones.📋

### Dispositivos de entrada y salida:⌨️🖱️🖨️

- Entrada: Permiten que los datos o el programa se lean en memoria principal desde los dispositivos de entrada o almacenamiento secundario bajo el control de instrucciones de entrada en la CPU.📥

- Salida: Muestran la información que ve el usuario.🖥️

# Harvard VS Von Neumann.

La arquitectura de Von Neumann y la arquitectura Harvard son los dos modelos teóricos principales en los que se basan las computadoras, siendo la de Von Neumann la más utilizada en CPU, consolas y dispositivos móviles.📱💻

## Diferencias:

Ambas arquitecturas son muy solidas, no se puede decir que una rinde más que otra, estas arquitecturas durante años han estado presentes siendo la de Von Neumann la mas popular. 🌟

**Algunas diferencias son:**

- Mientras la memoria en la de Von Neumann es única para instrucciones y datos, la de harvard se divide en dos pozos, uno para datos y otro de instrucciones.💾📑🧱

- La de Von Neumann se usa para GPU y CPU, la de Harvard para Microcontroladores y cache, esto es así ya que al aumentar la cantidad de buses en harvard significaría aumentar el perímetro del procesador haciendolo más grande y caro.🖥️💸📏

# Limitantes de la arquitectura Von Neumann.

La arquitectura de Von Neumann, si bien es el diseño fundamental de la mayoría de computadoras modernas, presenta una limitación principal que afecta su rendimiento, conocida como el cuello de botella de Von Neumann.🚧
  
## Origen: Memoria única y bus compartido.

El problema surge directamente de la forma en que Von Neumann concibió la organización de sus tres bloques básicos, específicamente la relación entre la Unidad Central de Proceso (CPU) y la Memoria.🧠↔️💾

- Memoria Única: El diseño de Von Neumann utiliza un pozo de RAM único para almacenar tanto las instrucciones del programa como los datos numéricos.📑➕🔢

- Bus Compartido: Debido a esta umión de la memoria, la CPU y la memoria comparten el mismo bus de datos y direccionamiento.🚌

La consecuencia directa de compartir este recurso es que las instrucciones y los datos deben ser captados de la memoria de manera secuencial. Esto se conoce como una arquitectura secuencial en lugar de paralela.⏳

## Cuello de botella:

El problema ha incrementado con el tiempo porque las velocidades de los procesadores han aumentado mucho más rápido que las de la memoria RAM. Esta diferencia temporal en la cantidad de datos que debe pasar entre la CPU y la memoria puede ser significativa, lo que provoca que la CPU pueda estar sin hacer nada esperando la información que necesita para procesar. Cabe destacar que este problema se puede presentar tambien en la arquitectura Harvard.

## Solución:

Un mecanismo que ha funcinado es la famosa "Memoria Cache", ya que, se incluye una de estas entre la CPU y la memoria principal. Esta caché es más rápida que la RAM normal y almacena datos e instrucciones de uso frecuente.
