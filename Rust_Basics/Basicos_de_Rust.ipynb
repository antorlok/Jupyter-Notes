# Básicos de Rust.

## ¿Qué es Rust y para que sirve hoy en día?

Rust es un lenguaje de programación de sistemas enfocado en la seguridad de memoria y el alto rendimiento. Su mayor innovación es que no tiene recolector de basura; en su lugar, utiliza un sistema de propiedad (ownership) que evita errores en tiempo de ejecución (como el acceso a memoria liberada o data races).

- ### ¿Para que es ideal?

    Sistemas operativos, motores de bases de datos, aplicaciones embebidas, WebAssembly y sistemas donde el rendimiento y la seguridad son críticos.

- ### ¿Empresas reconocidas que lo utilizan?
    Empresas que lo usan: Mozilla, Microsoft, Amazon Web Services (AWS), Discord, Cloudflare.

## Metodos de instalación.

- ### En entornos de linux.
    La forma estándar y recomendada (independientemente de la distribución) es usar rustup, el gestor de herramientas de Rust:
    
    **curl --proto '=https' --tlsv1.2 -sSf https://sh.rustup.rs | sh**
- ### En Windows.
    Instalar Rust en Windows tiene un pequeño requisito previo muy importante: necesita las herramientas de compilación de C++ para poder enlazar (linkear) los archivos binarios correctamente.

**Paso a paso:**

- Requisito previo (MSVC): Rust en Windows depende de las Visual Studio C++ Build Tools. Si no las tienes, el propio instalador de Rust te avisará y te dará la opción de descargarlas automáticamente, pero es bueno tenerlo en cuenta.

- Descarga de rustup: Ve a la página oficial de Rust (https://rustup.rs/) y haz clic en el botón para descargar rustup-init.exe. Esta es la herramienta oficial que gestiona las versiones de Rust.

- Ejecución en terminal: Al abrir el archivo rustup-init.exe, se abrirá una ventana de consola negra.

- Si te falta el requisito de C++, te preguntará si deseas instalarlo. Escribe y (sí).

- Una vez cubierto el requisito, verás un menú con opciones. Simplemente presiona la tecla 1 (Proceed with installation - default) y luego Enter.

- Finalización: El instalador descargará el compilador (rustc), el gestor de paquetes (cargo) y configurará tu PATH.

- Verificación: Cierra esa ventana, abre una nueva terminal (PowerShell o CMD) y comprueba las herramientas principales con estos comandos:
    - Verifica el compilador: **rustc --version**

    - Verifica el gestor de paquetes y proyectos: **cargo --version**

## Variables y Mutabilidad

    El enfoque de Rust es radicalmente distinto: las variables son inmutables por defecto para garantizar la seguridad. Si deseas modificar el valor de una variable posteriormente, debes usar explícitamente la palabra clave mut.

```rust
fn main() {
    // Variable inmutable (fija)
    let nombre = "Andres"; 
    
    // Variable mutable y anotación de tipo (i32 es entero de 32 bits)
    let mut edad: i32 = 25; 
    
    // Podemos cambiarla porque tiene 'mut'
    edad += 1; 
    
    println!("Nombre: {}, Edad: {}", nombre, edad);
}

main()
```

**Nota:** Se invoca al main() en el codigo para que se muestre en este NoteBook pero no se debe colocar

### Profundizando:

    Para observar detalladamente cada tipo de variable, a continuacion codigo Rust con comentarios explicando los tipos.

```rust
// =====================================================================
// DEFINICIONES PREVIAS (Para mostrar tipos personalizados)
// =====================================================================

// Un Struct agrupa datos relacionados bajo un mismo nombre.
struct Usuario {
    nombre: String,
    edad: u8,
}

// Un Enum permite representar múltiples variantes de datos.
enum Rol {
    Administrador,
    Invitado,
}

// =====================================================================
// FUNCIÓN PRINCIPAL: CATÁLOGO DE VARIABLES EN RUST
// =====================================================================

fn main() {
    // -----------------------------------------------------------------
    // 1. TIPOS ESCALARES (Representan un único valor)
    // -----------------------------------------------------------------

    // Enteros con signo (i) y sin signo (u). El número indica los bits.
    let edad: u8 = 25;               // De 0 a 255 (Ideal para edades).
    let saldo_bancario: i32 = -1500; // Puede ser negativo. i32 es el estándar.
    
    // Tipo de arquitectura: Su tamaño depende de tu PC (32 o 64 bits).
    let direccion_memoria: usize = 4096;

    // Flotantes (Decimales). f64 es el doble de preciso que f32.
    let temperatura: f32 = 28.5;
    let pi: f64 = 3.14159265359;

    // Booleanos (Verdadero o Falso)
    let es_estudiante: bool = true;

    // Caracteres (Ocupan 4 bytes, soportan cualquier símbolo Unicode)
    let letra_inicial: char = 'A';
    let icono: char = '🐧'; // Funciona perfecto.

    // -----------------------------------------------------------------
    // 2. TIPOS COMPUESTOS (Agrupan múltiples valores de tamaño fijo)
    // -----------------------------------------------------------------

    // Tuplas: Diferentes tipos de datos juntos. Tamaño inmutable.
    let coordenadas: (i32, f64, char) = (10, 45.2, 'N');
    let latitud = coordenadas.1; // Así accedemos al segundo valor (45.2).

    // Arreglos (Arrays): Mismo tipo de dato. Tamaño inmutable.
    let notas_semestre: [i32; 4] = [18, 20, 16, 19];

    // -----------------------------------------------------------------
    // 3. TEXTO (Cadenas de caracteres)
    // -----------------------------------------------------------------

    // String Slice (&str): Es una vista rápida a un texto fijo. No puede crecer.
    let lenguaje_favorito: &str = "Rust";

    // String: Un texto dinámico que se guarda en la memoria Heap. Puede modificarse.
    // Usamos 'mut' para que se le pueda agregar texto.
    let mut mensaje: String = String::from("Hola");
    mensaje.push_str(" Andres!"); 

    // -----------------------------------------------------------------
    // 4. REFERENCIAS Y PRÉSTAMOS (Borrowing)
    // -----------------------------------------------------------------

    // Referencia Inmutable: Permite leer datos sin tomar posesión.
    let ref_lectura: &String = &mensaje; 
    
    // Referencia Mutable: Permite modificar el dato original.
    // Ojo: Solo puede haber una referencia mutable a la vez.
    let mut contador = 0;
    let ref_escritura: &mut i32 = &mut contador;
    *ref_escritura += 1; // Sumamos 1 modificando directamente la memoria.

    // -----------------------------------------------------------------
    // 5. COLECCIONES DINÁMICAS (Crecen según necesites) 
    // -----------------------------------------------------------------

    // Vec<T> (Vector): Un array redimensionable.
    let mut lista_tareas: Vec<&str> = vec!["Instalar dependencias", "Configurar terminal"];
    lista_tareas.push("Compilar kernel"); 

    // -----------------------------------------------------------------
    // 6. MANEJO DE AUSENCIA Y ERRORES (Adiós al 'Null') 
    // -----------------------------------------------------------------

    // Option<T>: Representa algo que puede tener un valor (Some) o nada (None).
    let configuracion_guardada: Option<i32> = Some(1024);
    let sin_configuracion: Option<i32> = None;

    // Result<T, E>: Representa el éxito (Ok) o el fracaso (Err) de una operación.
    let lectura_exitosa: Result<&str, &str> = Ok("Archivo leído correctamente");
    let lectura_fallida: Result<&str, &str> = Err("Permiso denegado");

    // -----------------------------------------------------------------
    // 7. INSTANCIACIÓN DE TIPOS PERSONALIZADOS
    // -----------------------------------------------------------------

    // Creando una variable de tipo Struct
    let perfil = Usuario {
        nombre: String::from("Andres"),
        edad: 21,
    };

    // Creando una variable de tipo Enum
    let permiso_actual = Rol::Administrador;

    // Imprimimos algo sencillo al final para comprobar que corre
    println!("El catálogo de variables se compiló exitosamente. ¡Bienvenido a Rust, {}!", perfil.nombre);
}

main()
```

## Ownership y Borrowing a detalle
    El sistema de propiedad es la característica más distintiva de Rust.


- **Ownership (Propiedad):** Cada valor tiene un único dueño a la vez, y cuando el dueño sale del alcance (scope), el valor se libera automáticamente de la memoria.


- **Borrowing (Préstamo):** Para no tener que pasar la propiedad constantemente, Rust permite prestar valores mediante referencias. Existen referencias inmutables (&T, múltiples permitidas a la vez) y mutables (&mut T, solo una a la vez para prevenir condiciones de carrera).

```rust
fn calcular_longitud(s: &String) -> usize {
    s.len() // Solo lectura, no toma la propiedad del string 
}

fn agregar_texto(s: &mut String) {
    s.push_str(" - modificado"); // Permite modificar el dato original 
}

fn main() {
    let mut texto = String::from("Rust");
    
    // Préstamo inmutable
    let longitud = calcular_longitud(&texto); 
    println!("Longitud: {}", longitud);
    
    // Préstamo mutable
    agregar_texto(&mut texto); 
    println!("Texto final: {}", texto);
}

main()
```

## Condicionales y Ciclos
    Los condicionales no llevan paréntesis. Los bucles en Rust son más seguros y expresivos, utilizando rangos (ej. 0..5 que excluye el 5).

```rust
fn main() {
    let edad = 20;
    

    // 1. IF / ELSE IF / ELSE Tradicional [cite: 399, 400]
    if edad < 13 {
        println!("Es un niño");
    } else if edad < 18 {
        println!("Es un adolescente");
    } else {
        println!("Es un adulto"); // Se ejecuta este bloque
    }

    // 2. MATCH (El equivalente a 'switch', pero exhaustivo y seguro) 
    let dia = "martes";
    match dia {
        // Usamos el pipe '|' para evaluar múltiples opciones en una línea
        "lunes" | "martes" | "miercoles" | "jueves" | "viernes" => println!("Es un día laborable"),
        "sabado" | "domingo" => println!("Es fin de semana"),
        // El guion bajo '_' actúa como el 'default' atrapando cualquier otro valor
        _ => println!("Día no válido"),
    }

    // 3. IF COMO EXPRESIÓN (¡Buenas Prácticas!)
    // En lugar de crear una variable vacía y llenarla dentro de un if/else,
    // Rust nos permite asignar el resultado del 'if' directamente a la variable.
    // Nota: ¡No se pone punto y coma dentro de las llaves para que retorne el valor!
    let categoria = if edad >= 18 {
        "Mayor de edad"
    } else {
        "Menor de edad"
    };

    println!("La persona es: {}", categoria);
    
    // Bucle for con rango (imprime 0, 1, 2)
    for i in 0..3 {
        println!("Iteración: {}", i);
    }

    // Bucle While

    // 1. OBLIGATORIO: La variable de control debe ser mutable (mut).
    // Si fuera inmutable, no podríamos cambiar su valor y crearíamos un ciclo infinito.
    let mut bateria: u8 = 100;

    while bateria > 20 {
        println!("🔋 Batería al {}%. El sistema funciona en estado óptimo.", bateria);
    
        // 3. ACTUALIZACIÓN: Reducimos la batería en cada iteración. 
        // Rust soporta operadores de asignación compuesta como -= o +=
        bateria -= 25;
    }
}

main()
```

## Funciones
    La firma de la función requiere definir estrictamente los tipos. La flecha -> indica el tipo de retorno. Un detalle clave de diseño: en Rust, si omites el punto y coma (;) en la última expresión de una función, esta retornará ese valor de forma implícita.

```rust
// Retorno implícito (no lleva 'return' ni ';')
fn sumar(a: i32, b: i32) -> i32 {
    a + b 
}

fn main() {
    let resultado = sumar(10, 5);
    println!("La suma es: {}", resultado);
}

main()
```

## Nombramiento de variables

Rust tiene convenciones de estilo muy marcadas, definidas en su RFC 430. De hecho, si no las sigues, el compilador emitirá advertencias (warnings) sugiriendo que cambies el nombre para cumplir con el estándar.

- Variables, Funciones, Módulos y Archivos: Se usa estrictamente snake_case (todo en minúsculas separado por guiones bajos). Ejemplos del propio lenguaje son calcular_longitud o leer_archivo.
- Structs, Enums y Traits (Interfaces): Se usa estrictamente PascalCase (UpperCamelCase). Ejemplos claros son estructuras como Estudiante o enums como Estado.
- Constantes y variables estáticas: Se usa SCREAMING_SNAKE_CASE (todo en mayúsculas separado por guiones bajos).

Ejemplo:

```rust
// SCREAMING_SNAKE_CASE: Para constantes globales
const LIMITE_CONEXIONES: u32 = 100;

// PascalCase: Para definir un Struct o Enum 
struct ServidorWeb {
    // snake_case: Para los campos dentro del struct
    direccion_ip: String,
    puerto_activo: u16,
}

// PascalCase: Implementación del struct
impl ServidorWeb {
    // snake_case: Para métodos y funciones 
    // La palabra clave 'pub' es lo que hace pública la función en Rust, no la mayúscula.
    pub fn iniciar_servidor(ip: String, puerto: u16) -> Self {
        // snake_case: Variables locales
        let nuevo_servidor = ServidorWeb {
            direccion_ip: ip,
            puerto_activo: puerto,
        };
        nuevo_servidor
    }
}

// snake_case: Función principal
fn main() {
    // snake_case: Instancia de variable
    let mi_servidor = ServidorWeb::iniciar_servidor(String::from("127.0.0.1"), 8080);
    
    println!("Servidor en puerto: {}", mi_servidor.puerto_activo);
    println!("Límite del sistema: {}", LIMITE_CONEXIONES);
}

main()
```
