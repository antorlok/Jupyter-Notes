# Básicos de Go(golang)

## ¿Qué es y para que sirve?

    Go fue creado en Google con un objetivo claro: priorizar la simplicidad y la productividad de los desarrolladores. Es un lenguaje compilado, estáticamente tipado y destaca por tener la concurrencia integrada de forma nativa (mediante goroutines y canales). Incluye un recolector de basura (garbage collector) para la gestión automática de la memoria.

- ### ¿Para que es ideal?    
          Microservicios, APIs REST, herramientas de línea de comandos (CLI), aplicaciones en la nube y sistemas distribuidos.
- ### Empresas que lo usan.
          Empresas que lo usan: Google, Uber, Netflix, Twitch, Docker.

## Instalación.

- ### En entornos Linux.
    En diferentes distribuciones de preferencia, puedes instalarlo fácilmente desde la terminal:

    RedHAt Enterprise y derivados: sudo dnf install golang

    Debian GNU/Linux y derivados: sudo apt install golang
- ### En Windows.
    A diferencia de los gestores de paquetes en Linux, la forma oficial y más segura de instalar Go en Windows es mediante su instalador gráfico.

    **Paso a paso:**

    - Descarga oficial: Dirígete a la página de descargas oficial de Go (https://go.dev/dl/) y selecciona el instalador para Windows (usualmente un archivo con extensión .msi).

    - Ejecución del instalador: Abre el archivo descargado y sigue las instrucciones del asistente. Por defecto, Go se instalará en el directorio C:\Program Files\Go.

    - Variables de entorno: El instalador es bastante inteligente y configurará automáticamente la variable de entorno PATH por ti, lo que te permitirá ejecutar Go desde cualquier ubicación.

     - Verificación: Para comprobar que la instalación fue un éxito, abre una nueva ventana de terminal (puede ser PowerShell o el Símbolo del sistema / CMD) y escribe el siguiente comando:
 
        **go version**

## Nuestro Primer "Hola Mundo!!"

    En Go, todo el código ejecutable debe pertenecer al paquete main y tener una función main. Importamos el paquete fmt (format) para imprimir en consola. No es necesario colocar punto y coma (;) al final de las sentencias.

```go
package main

import "fmt"

func main() {
    fmt.Println("¡Hola mundo desde Go!")
}
```

## Variables y Tipos

    Go permite la declaración explícita de tipos usando la palabra reservada var y una declaración corta con inferencia de tipos usando :=. Todas las variables en Go son mutables por defecto, lo que significa que puedes reasignarles valores sin marcar explícitamente la mutabilidad.

```go
package main

import "fmt"

// =====================================================================
// DEFINICIONES PREVIAS (Fuera del main)
// =====================================================================

// En Go definimos los Structs usando la convención PascalCase para que
// sean públicos (exportados) o camelCase si queremos que sean privados.
type Usuario struct {
	Nombre string
	Edad   uint8
}

// =====================================================================
// FUNCIÓN PRINCIPAL: CATÁLOGO DE VARIABLES EN GO
// =====================================================================

func main() {
	// -----------------------------------------------------------------
	// 1. DECLARACIONES BÁSICAS Y MUTABILIDAD
	// -----------------------------------------------------------------
    // En Go las variables son mutables por defecto.
	
    // Declaración explícita con 'var'. Ideal cuando declaras sin inicializar.
	var lenguaje string = "Go"
	
	// Declaración corta ':='. El compilador infiere el tipo.
    // Es la forma más idiomática y recomendada dentro de funciones.
	version := 1.22
	esCompilado := true

	// -----------------------------------------------------------------
	// 2. TIPOS ESCALARES NUMÉRICOS (Enteros y Flotantes)
	// -----------------------------------------------------------------
	
    // 'int' estándar: su tamaño (32 o 64 bits) depende de la arquitectura de tu PC.
	poblacion := 8000000 
	
	// Tipos de tamaño fijo: Ocupan exactamente lo que dicen, igual que en Rust.
	var edadExacta uint8 = 21   // Entero sin signo (0 a 255)
	var saldoCuenta int32 = -50 // Entero con signo 
	
	// Flotantes: 'float64' es el estándar recomendado para cálculos de alta precisión.
	var temperatura float64 = 36.5

	// -----------------------------------------------------------------
	// 3. TEXTO Y CARACTERES (Rune y Byte)
	// -----------------------------------------------------------------
	
    // 'byte' es simplemente un alias de uint8. Representa un valor ASCII.
	var inicial byte = 'A'
	
	// 'rune' es un alias de int32. Representa un caracter Unicode (como el char de Rust).
	var icono rune = '🐧'

	// -----------------------------------------------------------------
	// 4. COLECCIONES: ARRAYS Y SLICES
	// -----------------------------------------------------------------
	
    // Arreglo (Array): Tamaño fijo en memoria. No puede crecer.
	var puertosFijos [3]int = [3]int{80, 443, 8080}
	
	// Slice: Es dinámico. Crece según necesites.
	tareas := []string{"Instalar Arch Linux", "Configurar Terminal"}
	tareas = append(tareas, "Aprender Go") // Así añadimos elementos

	// -----------------------------------------------------------------
	// 5. MAPAS (Diccionarios Hash)
	// -----------------------------------------------------------------
	
    // Mapea una clave de un tipo a un valor de otro tipo.
	codigosHTTP := map[string]int{
		"OK":        200,
		"Not Found": 404,
	}

	// -----------------------------------------------------------------
	// 6. PUNTEROS (Referencias directas a la memoria)
	// -----------------------------------------------------------------
	
    // Go tiene punteros (como C o C++), pero no permite aritmética de punteros por seguridad.
	var contador int = 10
	var punteroContador *int = &contador // Guardamos la dirección de memoria de 'contador'
	*punteroContador = 11                // Modificamos el valor original usando el puntero

	// -----------------------------------------------------------------
	// 7. INSTANCIACIÓN DE TIPOS PERSONALIZADOS
	// -----------------------------------------------------------------
	perfil := Usuario{
		Nombre: "Andres",
		Edad:   edadExacta,
	}

	// =====================================================================
	// IMPRESIÓN OBLIGATORIA (Para que el compilador no arroje error)
	// =====================================================================
	
	fmt.Println("=== Tipos Básicos ===")
	fmt.Printf("Lenguaje: %s (v%.2f) | Compilado: %t\n", lenguaje, version, esCompilado)
	fmt.Printf("Población: %d | Saldo: %d | Temperatura: %.1f\n", poblacion, saldoCuenta, temperatura)
	
	fmt.Println("\n=== Caracteres ===")
	// Usamos %c para que imprima el caracter visual y no el número del código
	fmt.Printf("Inicial (byte): %c | Icono (rune): %c\n", inicial, icono)

	fmt.Println("\n=== Colecciones ===")
	fmt.Println("Puertos (Array):", puertosFijos)
	fmt.Println("Tareas (Slice):", tareas)
	fmt.Println("Códigos HTTP (Mapa):", codigosHTTP)

	fmt.Println("\n=== Punteros y Structs ===")
	fmt.Println("Contador modificado vía puntero:", contador)
	fmt.Printf("Usuario Registrado: %s, Edad: %d\n", perfil.Nombre, perfil.Edad)
}
```

## Condicionales

    Los condicionales en Go son limpios: no requieren paréntesis envolviendo la condición y las llaves {} son obligatorias.

```go
package main

import "fmt"

func main() {
    edad := 20
    
    // 1. IF / ELSE IF / ELSE Tradicional [cite: 396, 397]
	if edad < 13 {
		fmt.Println("Es un niño")
	} else if edad < 18 {
		fmt.Println("Es un adolescente")
	} else {
		fmt.Println("Es un adulto") // Se ejecuta este bloque
	}

	// 2. SWITCH (El reemplazo limpio para evitar muchos "else if")
	// Go evalúa los casos de arriba hacia abajo y se detiene automáticamente.
	dia := "martes"
	switch dia {
	case "lunes", "martes", "miercoles", "jueves", "viernes":
		fmt.Println("Es un día laborable")
	case "sabado", "domingo":
		fmt.Println("Es fin de semana")
	default:
		fmt.Println("Día no válido")
	}

	// 3. IF CON INICIALIZACIÓN (¡Buenas Prácticas!)
	// Declaramos 'nota', le asignamos 85, y la evaluamos, todo en la misma línea.
	// La ventaja: la variable 'nota' se destruye al terminar el 'if', ahorrando memoria.
	if nota := 85; nota >= 80 {
		fmt.Printf("Aprobado con honores. Nota: %d\n", nota)
	}
}
```

## Ciclos 
    Go simplificó la sintaxis eliminando el while. El único bucle que existe es el for, el cual tiene estilo C y se puede usar de múltiples formas (como iterador o como condición simple).

```go
package main

import "fmt"

func main() {
    // For tradicional
    for i := 0; i < 3; i++ {
        fmt.Println("Iteración:", i)
    }
    
    // For actuando como un 'while'
    contador := 0
    for contador < 2 {
        fmt.Println("Contador:", contador)
        contador++
    }
}
```

## Funciones

    Las funciones en Go declaran el tipo de parámetro después del nombre de la variable, y el tipo de retorno se ubica al final, justo antes de abrir las llaves. Go permite retornar múltiples valores, una práctica común para el manejo de errores.

```go
package main

import "fmt"

// Función que recibe dos enteros y retorna un entero
func sumar(a int, b int) int {
    return a + b
}

func main() {
    resultado := sumar(10, 5)
    fmt.Println("La suma es:", resultado)
}
```

## Nombramiento de Variables

    En Go, el estándar oficial se inclina fuertemente por el uso de camelCase (primera letra minúscula, siguientes palabras con mayúscula) y PascalCase (todas las palabras inician con mayúscula). Go evita por completo el uso de guiones bajos (snake_case).

Pero aquí viene el dato más importante de Go: la capitalización determina la visibilidad (exportación) del elemento. No existen palabras reservadas como public o private.

- Variables, funciones y métodos privados (solo visibles dentro de su paquete): Se usa camelCase.

- Variables, funciones y métodos públicos (exportados para usarse en otros paquetes): Se usa PascalCase.

- Paquetes: Nombres cortos, de una sola palabra y todo en minúsculas (ej. fmt, http).

```go
package main

import "fmt"

// PascalCase: Esta estructura es PÚBLICA (exportada)
type UsuarioSistema struct {
    // PascalCase: Estos campos son PÚBLICOS
    NombreUsuario string
    Edad          int
    
    // camelCase: Este campo es PRIVADO (no accesible desde otros paquetes)
    tokenSesion   string
}

// PascalCase: Función PÚBLICA
func CrearUsuario(nombre string, edad int) UsuarioSistema {
    // camelCase: Variable local PRIVADA
    nuevoUsuario := UsuarioSistema{
        NombreUsuario: nombre,
        Edad:          edad,
        tokenSesion:   "abc123xyz",
    }
    return nuevoUsuario
}

// camelCase: Función PRIVADA
func calcularPrivilegios() {
    // Lógica interna...
}

func main() {
    usuario := CrearUsuario("Andres", 25)
    fmt.Println("Usuario creado:", usuario.NombreUsuario)
}
```

## Goroutines.

En la programación tradicional, si llamas a una función que tarda 5 segundos en ejecutarse, tu programa se congela esos 5 segundos esperando la respuesta antes de pasar a la siguiente línea de código.

**¿Qué es una Goroutine?**
Una Goroutine es simplemente una función que se ejecuta de forma independiente y simultánea (concurrente) al resto del programa.

- Son extremadamente ligeras: No son hilos (threads) del sistema operativo. Son gestionadas internamente por el propio entorno de ejecución de Go (Go Runtime).

- Cuestan muy poca memoria: Mientras un hilo tradicional de C++ o Java puede consumir Megabytes de RAM solo en arrancar, una Goroutine nace pesando apenas unos 2 Kilobytes. ¡Puedes tener cientos de miles de goroutines ejecutándose al mismo tiempo en una laptop común!

- Sintaxis ridículamente sencilla: Solo necesitas poner la palabra reservada go antes de llamar a cualquier función.


### Para este NoteBook.

Si la función principal (main) termina de leer todo el código, el programa se apaga de inmediato y asesina sin piedad a todas las goroutines que seguían trabajando de fondo. Para que el código funcione en esta celda y se puedan los resultados, debemos usar un mecanismo de sincronización llamado sync.WaitGroup, que básicamente le dice al programa: "Espera a que todas las tareas de fondo terminen antes de apagarte".

Código para este NoteBook:

```go
package main

import (
	"fmt"
	"sync"
	"time"
)

// =====================================================================
// FUNCIÓN CONCURRENTE
// =====================================================================

// Esta es la función que ejecutaremos en segundo plano.
// Recibe un ID, el tiempo que va a tardar, y un puntero al WaitGroup para avisar cuando termine.
func descargarArchivo(id int, segundos time.Duration, wg *sync.WaitGroup) {
	// 'defer' asegura que wg.Done() se ejecute justo antes de que la función termine,
	// sin importar si hubo errores o múltiples salidas. Es una excelente práctica en Go.
	defer wg.Done()

	fmt.Printf("⬇️ [Goroutine %d] Iniciando descarga...\n", id)
	
	// Simulamos el tiempo que tarda la descarga
	time.Sleep(segundos * time.Second)
	
	fmt.Printf("✅ [Goroutine %d] Descarga completada tras %d segundos.\n", id, segundos)
}

// =====================================================================
// FUNCIÓN PRINCIPAL
// =====================================================================

func main() {
	// 1. Instanciamos el WaitGroup
	var wg sync.WaitGroup

	fmt.Println("🚀 Iniciando el gestor de descargas...")
	inicio := time.Now()

	// 2. Lanzamos 3 tareas al mismo tiempo (Goroutines)
	// Como lanzamos 3, le sumamos 3 al contador del WaitGroup.
	wg.Add(3)

	// Fíjate en la palabra clave 'go'. ¡Eso es todo lo que se necesita!
	go descargarArchivo(1, 3, &wg) // Tarda 3 segundos
	go descargarArchivo(2, 1, &wg) // Tarda 1 segundo
	go descargarArchivo(3, 2, &wg) // Tarda 2 segundos

	// 3. Esperamos a que terminen
	fmt.Println("⏳ Tareas lanzadas a trabajar en el fondo. El programa principal está esperando...")
	wg.Wait() // Pausa la ejecución aquí hasta que el contador de wg llegue a 0.

	// 4. Calculamos el tiempo total
	tiempoTotal := time.Since(inicio)
	
	// Si esto no fuera concurrente, tardaría 6 segundos (3 + 1 + 2).
	// Como se ejecutan al mismo tiempo, solo tardará lo que tarde la más lenta (3 segundos).
	fmt.Printf("🎉 Todas las descargas finalizaron. Tiempo total real: %v\n", tiempoTotal)
}
```
